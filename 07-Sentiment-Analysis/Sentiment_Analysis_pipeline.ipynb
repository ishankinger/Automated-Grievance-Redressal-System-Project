```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/feedback_input.csv')
```

```python
import pickle

model_path = "model/sentiment_predict.pkl"
with open(model_path, 'rb') as f:
    loaded_svm_model = pickle.load(f)

df['predict_sentiment'] = loaded_svm_model.predict(df['feedback_text_processed'])
```

```python
df.to_csv('../data/feedback_output.csv', index=False)
```
