```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/06_input_time_prediction.csv')
```

```python
import pickle

model_path = "model/time_predict.pkl"
with open(model_path, 'rb') as f:
    loaded_svm_model = pickle.load(f)

cat_features = ['org_code', 'sex', 'state']
num_features = ['recvd_month']
text_main = 'grievance_text_processed'

df['predicted_days'] = loaded_svm_model.predict(df[[text_main] + cat_features + num_features])
```

```python
df.to_csv('../data/07_output.csv', index=False)
```
