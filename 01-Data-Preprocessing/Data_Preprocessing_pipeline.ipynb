```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/01_input_data.csv')
```

```python
df_classification = df.copy()
df_classification['grievance_text_processed'] = df_classification['grievance_text']
```

```python
df_classification['grievance_text_processed'] = df_classification['grievance_text_processed'].str.lower()
```

```python
import re

def remove_html_tags(text):
    if pd.isna(text):
        return text
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

df_classification['grievance_text_processed'] = df_classification['grievance_text_processed'].apply(remove_html_tags)
```

```python
import re

def remove_url(text):
    if pd.isna(text): 
        return text
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

df_classification['grievance_text_processed'] = df_classification['grievance_text_processed'].apply(remove_url)
```

```python
import string
exclude = string.punctuation

def remove_punc_better(text):
    if pd.isna(text):
        return text
    return ''.join(' ' if ch in exclude else ch for ch in text)

df_classification['grievance_text_processed'] = df_classification['grievance_text_processed'].apply(remove_punc_better)
```

```python
from nltk.corpus import stopwords
import nltk

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if pd.isna(text):
        return text
    words = [word for word in text.split() if word not in stop_words]
    return " ".join(words)

df_classification['grievance_text_processed'] = df_classification['grievance_text_processed'].apply(remove_stopwords)
```

```python
df_classification.to_csv("../data/02_input_anomaly_detection.csv", index=False)
```
