```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/05_input_risk_scoring.csv')
```

```python
import pickle

model_path = "model/urgency_model_svc.pkl"
with open(model_path, 'rb') as f:
    loaded_svm_model = pickle.load(f)

df['recvd_date'] = pd.to_datetime(df['recvd_date'], errors='coerce')
df['recvd_month'] = df['recvd_date'].dt.month.fillna(-1).astype(str)
df['recvd_day'] = df['recvd_date'].dt.dayofweek.fillna(-1).astype(str)

text_feature = 'grievance_text_processed'
categorical_features = ['org_code', 'sex', 'state', 'recvd_month', 'recvd_day']

df['urgency_tag'] = loaded_svm_model.predict(df[[text_feature] + categorical_features])
```

```python
df.to_csv('../data/06_input_time_prediction.csv', index=False)
```
