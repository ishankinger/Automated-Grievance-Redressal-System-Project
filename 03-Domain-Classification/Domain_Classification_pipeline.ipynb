```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/03_input_domain_classification.csv")
```

```python
import pickle

model_path = "model/svm_domain_classifier.pkl"
with open(model_path, 'rb') as f:
    loaded_svm_model = pickle.load(f)

df['org_code'] = loaded_svm_model.predict(df['grievance_text_processed'])
```

```python
output_csv_path = "../data/04_input_keyword_extraction.csv"
df.to_csv(output_csv_path, index=False)
```
